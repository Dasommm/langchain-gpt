```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(temperature=0.1, streaming=True, callbacks=[StreamingStdOutCallbackHandler()])
# Streaming은 응답이 생성되는 과정을 보여준다. 전체 응답을 보기 위해 작업이 끝날때까지 기다릴 필요 없다.
# callback을 위와 같이 작성해주면 console에서 응답의 진행을 볼 수 있다.

chef_promot = ChatPromptTemplate.from_messages([
    ("system", "You are a world-class international chef. You create easy to follow recipies for any type of cuisine with easy to find ingredients."),
    ("human", "I want to cook {cuisine} food.")
])

chef_chain = chef_promot | chat
```

```python
veg_chef_promot = ChatPromptTemplate.from_messages([
    ("system","You are a vegetarian chef specialized on making traditional recipes vegetarian. You find alternative ingredients and explain their preparation. You don't radically modify the recipe. If there is no alternative for a food just say you don't know how to recipe it."),
    ("human","{recipe}")
])

veg_chain = veg_chef_promot | chat

final_chain = {"recipe" : chef_chain} | veg_chain
# chef_chain의 결과값을 recipe로 저장한 후에 veg_chain으로 전달한다.
# runnableMap이라고 부른다. 한 체인의 결과값을 다음 체인의 입력값으로 사용한다.

final_chain.invoke({
    "cuisine": "indian"
})

# 결과값
# AIMessage(content="To make this Chicken Tikka Masala recipe vegetarian, we can substitute the chicken with a plant-based alternative such as tofu or paneer. Here's how you can modify the recipe:\n\nIngredients:\n- 1 lb firm tofu or paneer, cut into bite-sized pieces\n- 1 cup plain yogurt (use plant-based yogurt for a vegan option)\n- 2 tbsp lemon juice\n- 2 tbsp vegetable oil\n- 1 large onion, finely chopped\n- 3 cloves garlic, minced\n- 1-inch piece of ginger, grated\n- 1 can (14 oz) crushed tomatoes\n- 2 tsp garam masala\n- 1 tsp ground cumin\n- 1 tsp ground coriander\n- 1 tsp paprika\n- 1/2 tsp turmeric\n- 1/2 tsp cayenne pepper (adjust to taste)\n- Salt and pepper to taste\n- 1/2 cup heavy cream (use plant-based cream for a vegan option)\n- Fresh cilantro, chopped (for garnish)\n\nInstructions:\n1. Follow the same marinating process as the original recipe, but instead of chicken, marinate the tofu or paneer in the yogurt and spice mixture. Cover and refrigerate for at least 1 hour.\n\n2. Preheat the oven and bake the marinated tofu or paneer on skewers as you would with the chicken.\n\n3. In the skillet, follow the same steps for sautéing the onion, garlic, and ginger. Add the spices and cook until fragrant.\n\n4. Add the crushed tomatoes and simmer as directed in the original recipe.\n\n5. Stir in the heavy cream and the baked tofu or paneer. Allow the flavors to meld together as you simmer the dish.\n\n6. Taste and adjust seasoning as needed, then serve the Vegetarian Tikka Masala over steamed rice, garnished with fresh cilantro.\n\nBy making these simple ingredient swaps, you can enjoy a flavorful and satisfying Vegetarian Tikka Masala that stays true to the essence of the traditional recipe. Enjoy your meatless Indian feast!")
```
