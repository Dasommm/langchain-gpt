## caching을 사용하면 LM(언어모델)의 응답을 저장할 수 있다.
예를들어, 채팅봇이 있고, 항상 똑같은 질문을 받는다면 계속 답변을 만들지 않고 캐싱을 이용하여 저장하여 사용할 수 있다.
똑같은 질문을 계속 받더라도 답변을 재사용하여 비용을 절약할 수 있다.

---
### 캐시 사용할 수 있는 모듈 
from langchain.globals import set_llm_cache
from langchain.cache import InMemoryCache 

---
### 디버그 모듈 
from langchain.globals import set_debug

### DB에 저장해서 사용하는 캐시 모듈
from langchain.cache import SQLiteCache


```python
from langchain.chat_models import ChatOpenAI
from langchain.callbacks import StreamingStdOutCallbackHandler
from langchain.globals import set_llm_cache, set_debug
from langchain.cache import InMemoryCache, SQLiteCache

# import 한 후에 이렇게 작성하면 모든 response가 메모리에 저장이 된다.
# restart하면 캐싱이 사라지기 때문에 다시 생성한다.
# set_llm_cache(InMemoryCache())
set_llm_cache(SQLiteCache("cache.db")) # 데이터베이스에 저장할 수도 있음. 이름은 임의로 저장할 수 있고, 실행시에 해당 파일이 자동으로 생성된다.
set_debug(True) # 현재 실행과정 등 추가적인 정보들을 제공한다.


chat = ChatOpenAI(
    temperature=0.1,
    # streaming=True,
    # callbacks=[
    #     StreamingStdOutCallbackHandler(),
    # ]
)
chat.predict("How do you make italian pasta")

```

```python
# 위에 사용한 질문을 아래에서 똑같이 했을때, 답변을 재사용한다. 이전에는 3.2초가 걸렸으나 재질문을 했을때에는 바로 답변이 나오고 내용도 똑같다.
chat.predict("How do you make italian pasta")
```
