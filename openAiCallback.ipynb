## 1. openAI 모델을 사용할때 우리가 지출하는 비용, 토큰 등을 확인하는 방법

Prompt Tokens: 27
Completion Tokens: 491
Successful Requests: 2
Total Cost (USD): $0.0010225
와 같은 정보를 얻을 수 있다.

```python
from langchain.chat_models import ChatOpenAI
from langchain.callbacks import get_openai_callback

chat = ChatOpenAI(
    temperature=0.1,
)


with get_openai_callback() as usage:
    a = chat.predict("What is the recipe for soju")
    b = chat.predict("What is the recipe for bread")
    print(a,b,"\n")
    print(usage)
```

## 2. Serialization : 사용하는 모델 정보 확인 & 변경하기

채팅 모델이 아닌 LLM으로 작업할 때 text-davinci-003을 사용하게 되는데 만약 gpt-3.5-turbo-16k같은 다른 모델을 사용하고 싶을때 설정을 통해 변경할 수 있다.
(from langchain.chat_models import ChatOpenAI이 아닌 from langchain.llms.openai import OpenAI을 사용할때)

```python
from langchain.chat_models import ChatOpenAI
from langchain.llms.openai import OpenAI

chat = ChatOpenAI(
    temperature=0.1,
    max_tokens=450,
    model="gpt-3.5-turbo-16k" # 여기에 모델을 작성한다.
)

chat.save("model.json") # 실행하면 이 이름으로 파일이 생성된다.


```

```python
from langchain.chat_models import ChatOpenAI
from langchain.llms.openai import OpenAI
from langchain.llms.loading import load_llm

chat.load_llm("model.json") # 저장된 모델의 configuration 파일에서 정보를 불러온다.

print(chat)
```
